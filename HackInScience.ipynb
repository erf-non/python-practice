```python
'''
Side by side
Write a function combining two strings so they are displayed side by side, separated by a column of pipe (|) symbols, like this:

Lorem ipsum dolor sit amet, consectetur|Lorem ipsum dolor sit amet, consectetur
adipiscing elit. Sed non risus.        |adipiscing elit. Sed non risus.        
Suspendisse lectus tortor, dignissim   |Suspendisse lectus tortor, dignissim   
sit amet, adipiscing nec, utilisez sed |sit amet, adipiscing nec, utilisez sed 
sin dolor.                             |sin dolor.     
'''
import textwrap
import itertools


def sidebyside(left, right, width=79):
    if width % 2 == 0:
        width -= 2
    else:
         width -= 1
    result = ''
    
    wrapped_L = textwrap.wrap(left, width = width // 2, break_long_words=True)
    wrapped_R = textwrap.wrap(right, width = width // 2, break_long_words=True)
    
    
    for L, R in itertools.zip_longest(wrapped_L, wrapped_R, fillvalue = ' '):
        result += (f"{L: <{width//2}}|{R: <{width//2}}\n")
    return result
```

```python
'''
Roman Numerals
'''

def to_roman_numeral(number):
    if not 1 <= number <= 3999:
        raise ValueError("Number out of range for Roman numerals (1 to 3999)")

    roman_numerals = {
        1000: 'M',
        900: 'CM',
        500: 'D',
        400: 'CD',
        100: 'C',
        90: 'XC',
        50: 'L',
        40: 'XL',
        10: 'X',
        9: 'IX',
        5: 'V',
        4: 'IV',
        1: 'I'
    }

    result = ''
    for value, numeral in sorted(roman_numerals.items(), key=lambda x: x[0], reverse=True):
        while number >= value:
            result += numeral
            number -= value

    return result   
```

```python
'''
Caesar Cypher
Caesar cipher is an antique cryptographic method making some information 
hidden to whom doesn't have the key to decrypt it.
'''


from string import ascii_lowercase, ascii_uppercase
low = list(ascii_lowercase)
up = list(ascii_uppercase)

from string import ascii_lowercase, ascii_uppercase
low = list(ascii_lowercase)
up = list(ascii_uppercase)

def caesar_cypher_encrypt(s, key):
    result = ''
    for i in s:
        if i in low:
            p = low.index(i)
            result  += low[(p + key % 26)%26]
        elif i in up:
            p = up.index(i)
            result  += up[(p + key % 26)%26]
        else:
            result += i
    return result


def caesar_cypher_decrypt(s, key):
    result = ''
    for i in s:
        if i in low:
            p = low.index(i)
            result  += low[(p - key % 26)%26]
        elif i in up:
            p = up.index(i)
            result  += up[(p - key % 26)%26]
        else:
            result += i
    return result
```

```python
'''
Longest Collatz sequence
The collatz sequence goes like this:

Start by any number greater than zero.
If your number is even, divide it by two.
If your number os odd, multiply by three and add one.
'''


def collatz_length(n):
    sequence = []
    while not n == 1:
        if n % 2 == 0:
             n /= 2
             sequence.append(n)
        else:
            n = n*3+1
            sequence.append(n)
    
    return (len(sequence))

collatz_length(1)
```
